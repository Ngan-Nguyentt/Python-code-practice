### Q3. Perform Data Cleaning and Feature Engineering as per requirements

```python
print("Ngan Nguyen - 041154207")

string_cols = ['hasYard', 'hasPool', 'isNewBuilt', 'hasStormProtector']

# Confirms all 4 string columns contain only Yes or No values
print("Unique values in string columns:")
for col in string_cols:
    print(f"- {col}: {df[col].unique()}")

# Converts string to numeric
for col in string_cols:
    df[col] = (df[col] == 'Yes').astype(int)
    
    print(f"Converted '{col}': Yes = 1, No = 0")
```

### Q4. Apply Linear regression model (must set the class column)

```python
print("Ngan Nguyen - 041154207\n")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# separate input features in x
x = df.drop('price', axis=1)

# store the target variable in y
y = df['price']

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# check size of each vector
print("Train/Test Shapes:")
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)

# Build Linear Regression model on Training set
model = LinearRegression()
lrmodel = model.fit(x_train, y_train)

# display coefficients
print("\nCoefficients:")
print(lrmodel.coef_)

# display intercept
print("\nIntercept:")
print(lrmodel.intercept_)
```

### Q5.

```python
print("Ngan Nguyen - 041154207\n")

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

results = []

# test sizes for 80:20, 70:30, 85:15
split_list = [0.20, 0.30, 0.15]

for test_size in split_list:
    # split the dataset
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=42)
    
    # train the model
    model = LinearRegression()
    model.fit(x_train, y_train)
    
    # predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    # Calculate MAE for both Train and Test sets
    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    
    # Calculate RMSE for both Train and Test sets
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    # Split label
    split_name = f"{int((1 - test_size) * 100)}:{int(test_size * 100)}"
    
    # Store results
    results.append([split_name, train_mae, test_mae, train_rmse, test_rmse])

# Finally create results table
results_df = pd.DataFrame(results,columns=["Split", "Train MAE", "Test MAE", "Train RMSE", "Test RMSE"])
print(results_df)

# best split based on lowest Test RMSE
best_split = results_df.loc[results_df["Test RMSE"].idxmin()]

print("\nThe best performance based on lowest Test RMSE:")
print(best_split)
```

### Q6. List 3 highest positive & negative influential factors along with their coefficients

```python
print("Ngan Nguyen - 041154207\n")

# Use 80:20 split model to build coefficient table
coef_df = pd.DataFrame({'Feature': x.columns,'Coefficient': lrmodel.coef_}).sort_values('Coefficient', ascending=False).reset_index(drop=True)
print("All Feature Coefficients:")
print(coef_df.to_string(index=False))

# List 3 highest positive influential factors along with their coefficients
top3_positive = coef_df.head(3)
print("\nTop 3 Highest Positive Influential Factors")
print(top3_positive.to_string(index=False))

# List 3 highest negative influential factors along with their coefficients
top3_negative = coef_df.tail(3)
print("\nTop 3 Highest Negative Influential Factors")
print(top3_negative.to_string(index=False))
```
