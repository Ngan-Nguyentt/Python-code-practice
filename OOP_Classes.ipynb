**Task 1: Create a Simple Class**

```python
#create a class named Student with a method named say_hello
class Student:
    def say_hello(self):
        print("Hello, I am a student!")

#create an object and call the method using an object
s1 = Student()
s1.say_hello()
```

**Task 2: Using the __init__ Constructor**

```python
#create a class named Student with Constructor method __init__ and a method named display
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display(self):
        print("Student detail: ", self.name, self.age,"years old")

#create an object and call the method using an object
s1 = Student("Ngan", 25)
s1.display()
```

**Task 3: Creating Multiple Objects**

```python
#create a class named Car with Constructor method __init__ and a method named show_details
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def show_details(self):
        print(self.brand, self.price)

#create 2 objects and call the method using an object
car1 = Car("Ford",65000)
car2 = Car("Porsche",100000)
car1.show_details()
car2.show_details()
```

**Task 4: Class with Methods**

```python
#create a class named Car with 3 methods: __init__, add, multiply
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def multiply(self):
        return self.num1 * self.num2

#create an object with 2 arguments and call the method using an object
result = Calculator(5,10)
print("Sum: ", result.add())
print("Multiplication: ", result.multiply())
```

**Task 5: Multiple Constructors Using Class Method**

@classmethod: used to create an object using the class, not an existing instance and often build alternative constructors

```python
#create a class named Student with 2 arguments
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # use @classmethod to an object with a class named info, then split the string by a dash
    @classmethod 
    def from_string(info, data):
        name, marks = data.split("-")
        return info(name, int(marks))

    # Method to display information
    def display(self):
        print("Student name: ",self.name)
        print("Marks: ", self.marks)
        
s1 = Student.from_string("Ngan-89")
s1.display()        
```
