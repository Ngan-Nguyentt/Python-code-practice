### Question 1 – Encapsulation

Scenario: A Bank Account system must protect balance.

```python
# Base class BankAccount
class BankAccount:
    def __init__(self, balance):
        # Private attribute for balance
        self.__balance = float(balance)
        
    # Increase balance by a positive amount
    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Deposit must be positive.")

    # Decrease balance if sufficient funds are available
    def withdraw(self,amount):
        if amount <= 0:
            print("Withdraw must be positive.")
        elif amount > self.__balance:
            print("Insufficient funds!")
        else:
            self.__balance -= amount
    
    # return the final balance
    def get_balance(self):
        return self.__balance

account = BankAccount(2000)
account.deposit(100)
account.withdraw(200)

print(f"Final balance: ${account.get_balance()}")
```

### Question 2 – Inheritance

Scenario: School Management System.


```python
# Base class Person
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

# Derived class Student
class Student(Person):
    def __init__(self,name,age,student_id):
        #Inheritance: call the parent class constructor
        super().__init__(name,age)
        self.student_id = student_id

    def display_info(self):
        print("Student name:", self.name)
        print("Student age:", self.age)
        print("Student ID:", self.student_id)

s1 = Student("Ngan Nguyen",25,"041154207")
s1.display_info()
```

### Question 3 – Polymorphism
Scenario: Different animal sounds.

```python
# Base class Animal
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Subclass Dog
class Dog(Animal):
    # Method override
    def sound(self):
        print("Dog sounds woof!")


# Subclass Cat
class Cat(Animal):
    # Method override
    def sound(self):
        print("Cat sounds meow!")

# Polymorphism
for a in [Dog(),Cat()]:
    a.sound()
```

### Question 4 – Protected Members
Scenario: Company employee data.

```python
# Base class Employee
class Employee:
    def __init__(self,name,salary):
        self.name = name
        # Protected attribute for salary
        self._salary = salary

# Subclass Manager
class Manager(Employee):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department = department

    # Access and display the protected salary from the subclass
    def display_salary(self):
        print(f"Manager: {self.name} \nSalary: ${self._salary} \nDepartment: {self.department}")

emp1 = Manager("Ngan Nguyen",78000,"IT")
emp1.display_salary()
```

### Question 5 – Combined OOP
Scenario: Vehicle Rental System.

```python
# Base class Vehicle
class Vehicle:
    def rent_price(self):
        return 0
        
# Subclass Car
class Car(Vehicle):
    # Method override
    def rent_price(self):
        return 100
        
# Subclass Bike
class Bike(Vehicle):
    # Method override
    def rent_price(self):
        return 15

# Create vehicle objects and demonstrate polymorphism
vehicles = [Car(), Bike()]

# Print prices
for v in vehicles:
    print(f"{type(v).__name__} rent price per day: $ {v.rent_price()}")
```
