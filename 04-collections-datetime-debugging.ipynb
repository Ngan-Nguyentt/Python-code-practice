### Question 1 

Write a Python program that:

Accepts 5 student scores from the user.

Stores them in a list.

Removes duplicates.

Sorts the list in descending order.

Displays the top 3 scores.

```python
# Create a list and accept 5 student scores
scores = []

for i in range(1, 6):
    score = float(input(f"Enter student score {i}: "))
    scores.append(score)
    
print("A list of 5 student score: ", scores)
```

```python
# Remove duplicates --> select distinct scores
scores = list(set(scores))
print(f"After removing duplicates: {scores}")

# Sorts the list in descending order
scores.sort(reverse=True)
print(f"Sorted (highest first): {scores}")

# Display top 3 scores
print("\nTop 3 Scores:")
for i in range(min(3, len(scores))):
    print(f"  {i + 1}. {scores[i]}")
```

### Question 3 - Datetime 


1. Explain the purpose of the datetime module in Python.
2. What is the difference between datetime.now() and datetime.today()?
3. Discuss a scenario where you might need to convert a string representation of a date into a datetime object.
4. Write a code to find the number of days remaining until your next birthday using datetime module functions.

1. The purpose of the datetime in Python is to help deal with timestamps in code

2. The difference between datetime.now() and datetime.today():

* datetime.now(): returns the current local date and time (and can include a time zone if provided).

* datetime.today(): also returns the current local date and time, but does not support time zone arguments

```python
from datetime import datetime, timezone

now = datetime.now()
now_utc = datetime.now(tz=timezone.utc)
today = datetime.today()

print("Now: ", now)
print("Now (UTC): ", now_utc)
print("Today: ",today)
```

3. A common scenario is processing dates from a CSV file or user input.
* Dates arrive as strings (e.g., "2001-10-06"), but strings can't be compared or used in calculations.
* strptime() converts the string into a datetime object


```python
date_str = "2001-10-06"
date_obj = datetime.strptime(date_str, "%Y-%m-%d")

print(date_obj)
print(date_obj.year)
```

4. Write a code to find the number of days remaining until your next birthday using datetime module functions.

```python
today = datetime.now()
birthday = datetime(today.year, 10, 6)

# If birthday already passed this year, use next year
if birthday < today:
    birthday = datetime(today.year + 1, 10, 6)

days_remaining = (birthday - today).days
print(f"Days until next birthday: {days_remaining}")
```

### 5. Debugging a Program

Given the buggy code:

```python
numbers = [10, 20, 30, 40]
total = 0

for i in range(5):
    total += numbers[i]

print("Total:", total)
```

Tasks:

Use pdb to find the error.

Fix the error.

Show the correct output.

```python
import pdb

numbers = [10, 20, 30, 40]
total = 0

for i in range(5):
    pdb.set_trace()
    total += numbers[i]

print("Total:", total)
```

* p i → print current value of i
* p len(numbers) → print list length (shows 4)
* p numbers[i] → print element at index i (crashes when i = 4)
* n → next line
* s → step into function
* c → continue to next breakpoint
* q → quit debugger
